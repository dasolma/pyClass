```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_olivetti_faces
faces = fetch_olivetti_faces()
print len(faces.images)
fig = plt.figure(figsize=(16, 16))
for i in range(400):
    plt.subplot(20, 30, i + 1) # It starts with one
    plt.imshow(faces.images[i], cmap=plt.cm.gray)
    plt.grid(False);
    plt.title("%d"%(i+1))
    plt.xticks([]);
    plt.yticks([]);
```

```python
y_faces = [0] * 400
y_faces[10:19] = [1] * 10
y_faces[32] = 1; y_faces[37] = 1; y_faces[38] = 1;
y_faces[50:59] = [1] * 10
y_faces[63] = 1; y_faces[64] = 1; y_faces[69] = 1;
y_faces[120] = 1; y_faces[121] = 1; 
y_faces[124:139] = [1] * 16
y_faces[160] = 1; y_faces[161] = 1; 
y_faces[164:169] = [1] * 6
y_faces[180] = 1; y_faces[181] = 1; y_faces[182] = 1; y_faces[185] = 1;  
y_faces[189:192] = [1] * 4
y_faces[194] = 1;
y_faces[196:199] = [1] * 4
y_faces[260:279] = [1] * 20
y_faces[300:309] = [1] * 10
y_faces[330:339] = [1] * 10
y_faces[358:369] = [1] * 12

y_faces = y_faces[:400]

fig = plt.figure(figsize=(16, 16))
for i in range(400):
    if y_faces[i] == 1:
        plt.subplot(20, 30, i + 1) # It starts with one
        plt.imshow(faces.images[i], cmap=plt.cm.gray)
        plt.grid(False);
        plt.title("%d"%(i+1))
        plt.xticks([]);
        plt.yticks([]);
```

```python
def predict(x, y, pipeline, cv=5):
    from sklearn.pipeline import Pipeline
    from sklearn import cross_validation
    from sklearn import metrics

    model = Pipeline(pipeline)
    predicted = cross_validation.cross_val_predict(model, x, y, cv=cv)
    print metrics.accuracy_score(y, predicted)
    
    return model
    
X = faces.data
Y = y_faces
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size = 0.25, random_state=33)
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import StandardScaler


model = predict(X,Y,[ ('normalizador', StandardScaler()), ('modelolineal', SGDClassifier(average=True))], cv=100)
```

```python
#reduce dimensionality
from sklearn.decomposition import RandomizedPCA
n_eigenfaces = 25

pca = RandomizedPCA(n_components=n_eigenfaces, whiten=True)
features = pca.fit_transform(X)

model = predict(features,Y,[ ('normalizador', StandardScaler()), ('modelolineal', SGDClassifier(average=True))], cv=100)
```

```python
model.fit(X_train, y_train)
y_pred=model.predict(X_test)   

from sklearn import metrics

metrics.accuracy_score(y_test,y_pred)
```

```python
print metrics.classification_report(y_test,y_pred)
```

```python
print metrics.confusion_matrix(y_test,y_pred)
```
